```python
def inject_random_nan(data_nan,missing_rate=0.1): # Eksik veri eklenecek grup,kayıp oranı 0.1 => %10 kayıp oranı
    data_copy=data_nan.copy().astype(float) # veri kopyalama
    mask=np.random.random(data_nan.shape)<missing_rate # mask oluşturma
    data_copy[mask]=np.nan # belirtilen yerlere nan enjekte etme
    return data_copy,mask

```

```python
data_whit_nan,nan_mask =inject_random_nan(data_nan,missing_rate=0.15) #fonksiyonu %15 kayıpla dene
print(nan_mask)
print(data_whit_nan)
```

```python
# Her sutunun kendi ortalaması ile NaN verilerini doldurma
def NaN_doldur(data_nan):
    data_filled=data_nan.copy() # Data kopyalama
    col_means=np.nanmean(data_filled,axis=0) # kolon bazında axis=0 değeri kullanılır

    for idx in range(data_nan.shape[1]):  # .shape[idx]: n boyutlu bir dizinin belirtilen indeksdeki boyutunu getirir
        nan_mask=np.isnan(data_nan[:,idx]) # NaN olan yerleri  belirledik
        data_filled[nan_mask,idx]=col_means[idx] # NaN yerine o sütünün ortalaması ile doldurduk.

    return data_filled


```

```python
doldurulmus_data=NaN_doldur(data_nan)
print("Sütun ortalaması ile doldurulmuş data :")
print(doldurulmus_data)
```

```python
# Klasik standart sapma formülünün python döngüsü ile yapılması
def standart_sapma(veriler):
    ortalama = sum(veriler) / len(veriler)

    karelerin_farki = []
    for veri in veriler:
        karelerin_farki.append((veri - ortalama) ** 2)

    varyans = sum(karelerin_farki) / len(veriler)
    return math.sqrt(varyans)

```

```python
# Öklid Mesafesi Problemi
```

```python
# Saf python döngüsü
def oklid_mesafe(v1,v2):
    kare_toplam_fark=0
    for i in range(len(v1)):
        kare_toplam_fark+=(v1[i]-v2[i])**2
    return math.sqrt(kare_toplam_fark)
```

```python
# Numpy Kütüphanesi
def oklid_mesafe_numpy(v1,v2):
    return np.linalg.norm(np.array(v1)-np.array(v2))
```

```python
v1=[1,2,3]
v2=[4,5,6]
```
