```python
from cProfile import label

import numpy as np
import pandas as pd
df=pd.read_csv('train.csv')
```

```python
 # Hangi sütunlarda eksik değer var?
df.isna().sum()
```

```python
# Bağımsız değişkenler:
#  GrLivArea (yaşanabilir alan, metrekare)
#  OverallQual (evin genel kalite puanı)
# Bağımlı değişken:
#  SalePrice (evin satış fiyatı)

x=df[['GrLivArea','OverallQual']]
y=df['SalePrice']
```

```python
# Gerekirse seçtiğiniz sütunlardaki boş satırları atın.
df.isnull().sum()
```

```python
# scikit-learn kütüphanesinden LinearRegression kullanın
from sklearn.linear_model import LinearRegression
model=LinearRegression()
```

```python
# Modeli eğitin (fit).
model.fit(x,y)
```

```python
# Katsayıları (coef_, intercept_) ekrana yazdırın
print('Intercept',model.intercept_)
print('Coefficient',model.coef_[0])
print('Coefficient of Determination',model.coef_)
```

```python
print('1344 m^2 ve 7 genel kalite puanına sahip bir evin tahmini fiyati:',model.predict([[1344,7]])[0])
y_pred=model.predict(x)
y_true=y
```

```python
# R², MAE, RMSE hesaplayın ve yorumlayın.
from sklearn.metrics import r2_score,mean_absolute_error,mean_squared_error
import numpy as np
r2=r2_score(y_true,y_pred)
mea=mean_absolute_error(y_true,y_pred)
mse=mean_squared_error(y_true,y_pred)
rmse=np.sqrt(mse)
```

```python
print('R2 Score',r2)
print('Mean Absolute Error',mea)
print('Mean Squared Error',mse)
print('Root Mean Squared Error',rmse)
```

##  Model Başarılı Mı
- R2 skoru için bir inceleme yaparsak; modelimiz %71 başarı oranına sahip.Modelin iyi bir başarı durumuna sahip olduğunu söylemeiz yanlış olmaz.Gerçek dünya verileri ile karşılaştırdığımızda modelin genel ilişkileri iyi bir şekilde kavradığını görüyoruz.
- Ancak mean-absolute-error değeri açısından bakarsak +/-28756 gibi bir sapmanın olması küçük ölçekli bir emlak işinde doğru sonuçlar vermeyeceği oldukça açıktır.Özellikle Mean Absolute Error ile Root Mean Squared Error arasındaki farkın fazla olması veri grubunda 'outlier' aykırı değerlerin fazla olduğunu gösterir.Daha net ve başarılı bir sonuç almak istiyorsak bu aykırı değer grubunun hariç tutulması gerekir.Mesela normal dağılıma uydurmak için en uygun yöntem veri grubu aralığını medyandan her iki tarafa  3 standart sapma kadar aralıkta belirlemek hem verilerin büyük çoğunluğnun dahil edilmesini sağlar hem de daha yüksek doğruluk oranları alabilirz.
