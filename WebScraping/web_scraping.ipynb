```python
import undetected_chromedriver as uc
from time import sleep
import pandas as pd
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.common.by import By
from selenium.webdriver.support import expected_conditions as EC
```

```python
options=uc.ChromeOptions()
driver=uc.Chrome(options=options,version_main=142)
```

```python
driver.get('https://www.sahibinden.com/kiralik/isparta-merkez')
sleep(5)
```

```python
# Accept Cookies
cookie_btn=driver.find_element(By.XPATH,'//*[@id="onetrust-accept-btn-handler"]')
driver.execute_script('arguments[0].click();',cookie_btn)
sleep(3)
```

```python
# Her sayfada 50  sonuç göster
driver.execute_script('arguments[0].click();',WebDriverWait(driver,20).until(EC.element_to_be_clickable((By.XPATH,'//*[@id="searchResultsSearchForm"]/div/div[3]/div[4]/div[2]/ul/li[2]/a'))))
sleep(5)
```

```python
title_list=[]
price_list=[]
monthday_list=[]
year_list=[]
room_list=[]
m2_list=[]
loca_list=[]
page_number=1
```

```python
while True:

   # Ürün bilgileri
   item_titles=driver.find_elements(By.XPATH,'//*[@id="searchResultsTable"]/tbody/tr/td[2]/a[1]')
   item_m2=driver.find_elements(By.XPATH,'//*[@id="searchResultsTable"]/tbody/tr/td[3]')
   item_loca=driver.find_elements(By.XPATH,'//*[@id="searchResultsTable"]/tbody/tr/td[7]')
   item_prices=driver.find_elements(By.XPATH,'//*[@id="searchResultsTable"]/tbody/tr/td[5]/div/span')
   item_monthdays=driver.find_elements(By.XPATH,'//*[@id="searchResultsTable"]/tbody/tr/td[6]/span[1]')
   item_years=driver.find_elements(By.XPATH,'//*[@id="searchResultsTable"]/tbody/tr/td[6]/span[2]')
   item_rooms=driver.find_elements(By.XPATH,'//*[@id="searchResultsTable"]/tbody/tr/td[4]')



   # Döngü ile verileri listelere çekme
   for title in item_titles:
       title_list.append(title.text)
   for price in item_prices:
       price_list.append(price.text)
   for monthday in item_monthdays:
       monthday_list.append(monthday.text)
   for year in item_years:
       year_list.append(year.text)
   for room in item_rooms:
       room_list.append(room.text)
   for m2 in item_m2:
       m2_list.append(m2.text)
   for loca in item_loca:
       loca_list.append(loca.text)

   try:
       next_btn=driver.find_element(By.XPATH,'//a[@title="Sonraki"]')
       driver.execute_script('arguments[0].click();',next_btn)
       sleep(5)
       page_number+=1

   except Exception as e:
       print(e)
       print('Son sayfaya gelindi işlem bitmiştir.')
       break

driver.close()
```

```python
# Başlığı kelimelere ayırma
title_spt=[]
for title in title_list:
    title_spt.append(title.split())
```

```python
datas={
    'title':title_list,
    'price':price_list,
    'monthday':monthday_list,
    'year':year_list,
    'room':room_list,
    'm2':m2_list,
    'location':loca_list
}
```

```python
df=pd.DataFrame(datas)
df.to_csv('sahibinden_ısparta_kiralık.csv',index=False)
```

```python
# 'datas' sözlüğündeki her listenin uzunluğunu yazdır
for key, value in datas.items():
    print(f"{key}: {len(value)} adet veri var.")
```
